```python
from keras.models import load_model
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
x = load_model("checkpoint/model_val_loss-0.04.h5")
```

```python
def pred(mode, picture):
    if mode == 0:
        img = image.load_img("data/validation_set/non_hemmorhage_data/{0}".format(picture) , target_size = (299, 299))
        plt.imshow(img)
        img = image.img_to_array(img)
        img = np.expand_dims(img, axis = 0)
        img /= 255.
        prediction = x.predict(img)
    else:
        img = image.load_img("data/validation_set/hemmorhage_data/{0}".format(picture) , target_size = (299, 299))
        plt.imshow(img)
        img = image.img_to_array(img)
        img = np.expand_dims(img, axis = 0)
        img /= 255.

        prediction = x.predict(img)
        
    if prediction[0] > 0.5:
        print("Ei aivoverenvuotoa {:.2%} varmuudella".format(prediction[0][0]))
    else:
        print("Aivoverenvuoto {:.2%} varmuudella".format(1-prediction[0][0]))
    plt.show()
```

```python
pred(1, "aivoverenvuoto.jpg")
```

```python
for i in os.listdir("data/validation_set/hemmorhage_data"):
    pred(1, str(i))
```

```python
for i in os.listdir("data/validation_set/non_hemmorhage_data"):
    pred(0, str(i))
```
